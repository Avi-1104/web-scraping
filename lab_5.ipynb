```python
import numpy as np
import matplotlib.pyplot as plt

x = np.array(df['Sem 5']).reshape(-1, 1)
y = np.array(df['Sem 6']).reshape(-1, 1)


plt.scatter(x, y)


plt.xlabel('Semester 5')
plt.ylabel('Semester 6')
plt.title('Scatter Plot of Semester 5 vs Semester 6')


plt.show()
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25)
r=LinearRegression()
r.fit(x_train,y_train)
```

```python
y_predict= r.predict(x_test)
y_predict
```
